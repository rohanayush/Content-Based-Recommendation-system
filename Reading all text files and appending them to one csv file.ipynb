```python
#Giving path for our text files
path="D:/Machine Learning/bbc/bbc"
```

```python
folder = ["business","others","politics","sport","tech"]
```

```python
#Linking  path to os 
os.chdir(path)
```

```python
#Python empty lists creation with names x and y
x=[]
y=[]
```

```python
#Looping through our file structure inside path 
for i in folder:
    files = os.listdir(i)
    
    for text_files in files:
        file_path = i + "/" +text_files
        print ("reading file...",file_path)
        with open(file_path) as o:
            data = o.readlines()
        data = ' '.join(data)
        x.append(data)
        y.append(i) 
        
            
            
    
```

```python
x[0]
```

```python
data = {'news':x, 'type':y}
df = pd.DataFrame(data)
print('writing csv file...')
df.to_csv('d:/machine learning/bbc/bbc/class.csv', index = False)
```
